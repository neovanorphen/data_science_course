Carga de datos con OPEN

```python
data3 = open("../datasets/customer-churn-model/Customer Churn Model.txt","r")
```

```python
cols = data3.readline().strip().split(",")
n_cols  = len(cols)
n_cols
```

```python
counter = 0
main_dict = {}
for col in cols: 
    main_dict[col] = []
```

```python
for line in data3:
    values = line.strip().split(",")
    for i in range(len(cols)):
        main_dict[cols[i]].append(values[i])
    counter+=1
print ("el dataset tiene %d filas y %d columnas"%(counter,n_cols))
```

```python
df3 = pd.DataFrame(main_dict)
df3.head()
```

## Lectura y escritura de archivos

```python
infile = ("../datasets/customer-churn-model/Customer Churn Model.txt")
outfile = ("../datasets/customer-churn-model/customers with tabs.txt")

with open(infile, "r") as infile1:
    with open(outfile,"w") as outfile1:
        for line in infile1:
            fields = line.strip().split(",")
            outfile1.write("\t".join(fields)+"\n")
```

```python
df4 = pd.read_csv(outfile, sep = "\t" )
df4.head()
```
