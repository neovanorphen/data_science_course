# Simulación de Monte Carlo

* Generar 2 numeros aleatorios entre 0 y 1
    * calcular x*x + y*y
    * si el valor es menor a 1 esta dentro del circulom sino esta afuera

* Calculamos el numero total de veces que estan dentro del circulo y se divide entre el numero total de intentos
* usamos dicha probabilidad para aproximar el valor de pi
* repetir el experimento un numero x de veces para tener diferentes aproximaciones de pi
* Calcular el promedio de los x experimentos para obtener el valor final

```python
def pi_montecarlo(n, n_exp):
    pi_avg = 0
    pi_value_list = []
    
    for i in range(n_exp):
        value = 0
        x = np.random.uniform(0,1,n).tolist()
        y = np.random.uniform(0,1,n).tolist()
        for j in range(n):
            z = np.sqrt((x[j] * x[j]) + (y[j] * y[j]))
            if z <= 1:
                value += 1

        float_value = float(value)
        pi_value = float_value *4 / n
        pi_value_list.append(pi_value)
        pi_avg += pi_value

    pi = pi_avg/n_exp
    print(pi)
    fig = plt.plot(pi_value_list)
    return (pi,fig)
```

```python
pi_montecarlo(10000,200)
```
